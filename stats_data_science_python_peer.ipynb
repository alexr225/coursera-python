## Is there a significant difference in median value of houses bounded by the Charles river or not? (T-test for independent samples)

```python
# Hypothesis: There is no difference in median value of houses bounded by the charles river
out = ss.ttest_ind(boston_df['MEDV'], boston_df['CHAS'])
```

```python
ss.levene(boston_df['MEDV'], boston_df['CHAS'], center = 'mean')
```

```python
# Less than 0.05 so reject hypothesis
```

## Is there a difference in Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE)? (ANOVA)

```python
# Hypothesis: There is no difference in the proportion of median values of houses by Age
```

```python
# Test for equality of variance
ss.levene(boston_df[boston_df['age_group'] == "35 years and younger"]['MEDV'],
                   boston_df[boston_df['age_group'] == "Between 35 and 70 years"]['MEDV'], 
                   boston_df[boston_df['age_group'] == "70 years and older"]['MEDV'], 
                   center='mean')
```

```python
thirtyfive_lower = boston_df[boston_df['age_group'] == "35 years and younger"]['MEDV']
thirtyfive_seventy = boston_df[boston_df['age_group'] == "Between 35 and 70 years"]['MEDV']
seventy_older = boston_df[boston_df['age_group'] == "70 years and older"]['MEDV']
```

```python
ss.f_oneway(thirtyfive_lower, thirtyfive_seventy, seventy_older)
```

```python
# Less than 0.05 so reject hypothesis
```

## Can we conclude that there is no relationship between Nitric oxide concentrations and proportion of non-retail business acres per town? (Pearson Correlation)

```python
# Hypothesis: There is no correlation between NO concentrations and INDUS per town

ss.pearsonr(boston_df['NOX'], boston_df['INDUS'])
```

```python
# Less than 0.05 so reject null hypothesis, there is a correlation
```

## What is the impact of an additional weighted distance  to the five Boston employment centres on the median value of owner occupied homes? (Regression analysis)

```python
# Hypothesis: There is no impact on the median value of owner occupied homes relative to the five boston employment centres
```

```python
X = boston_df['DIS']
Y = boston_df['MEDV']

## add an intercept (beta_0) to our model
X = sm.add_constant(X) 

model = sm.OLS(Y, X).fit()
predictions = model.predict(X)

# Print out the statistics
model.summary()
```

```python
# p<0.05 so hypothesis is rejected as there is an impact
```
